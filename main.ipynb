```python
data=data[1:]
```

```python
years=[row[1] for row in data]
years
```

```python
year_counts={}
```

```python
for y in years:
    if y in year_counts:
        year_counts[y]+=1
    else:
        year_counts[y]=1
year_counts        
```

```python
import datetime
dates=[datetime.datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data]
```

```python
dates[:5]
```

```python
date_counts={}
```

```python
for d in dates:
    if d in date_counts:
        date_counts[d]+=1
    else:
        date_counts[d]=1
        
date_counts        
```

```python
def returnDic(rowno):
    dic={}
    for row in data:
        if row[rowno] in dic:
            dic[row[rowno]]+=1
        else:
            dic[row[rowno]]=1
    return dic
            
            
sex_counts=returnDic(5)
race_counts=returnDic(7)
```
